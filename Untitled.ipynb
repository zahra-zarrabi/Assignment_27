```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import imageio
import matplotlib.animation as animation
from matplotlib.animation import PillowWriter
import random
fig = plt.figure()
```

```python
image=cv2.imread('tv.jpg',cv2.IMREAD_GRAYSCALE)
roi=image[450:800,480:950]
plt.imshow(roi,cmap='gray')
```

```python
rows ,col= roi.shape
number=1
tv=[]
for n in range(30):
    for i in range(1,rows-1):
        for j in range(1,col-1):
            rnd=random.randint(0,5)
            if rnd<number:
                roi[i,j]=0
            elif rnd>number:
                roi[i,j]=255          
    image[450:800,480:950]=roi
    tv.append(image)
plt.imshow(image,cmap='gray')
#     plt.pause(2)
```

```python
# image[450:800,480:950]=roi

plt.imshow(image,cmap='gray')

# imageio.imwrite('tv.gif',tv)
ani=animation.ArtistAnimation(fig,tv,interval=50,blit=True,repeat_delay=500)
writer=PillowWriter(fps=20)
ani.save('de.gif',writer=writer)
plt.show()
```
